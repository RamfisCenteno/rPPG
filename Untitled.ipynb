```python
plt.figure(figsize=(14, 7), dpi=100)
x = data['BVP']
peaks, _ = find_peaks(x, height=.5, distance=15)
plt.plot(x)
plt.plot(peaks, x[peaks], "x")
plt.plot(np.zeros_like(x), "--", color="gray")
sns.set(font_scale =1.5, style = 'ticks' )
plt.show()
```

```python
y = pd.DataFrame(_, columns=['peak_heights'])
```

```python
y
```

```python
peaks = pd.DataFrame(peaks, columns=['peak_points'])
peaks['peak_seconds'] = peaks['peak_points']/7.43
peaks['diff'] = peaks['peak_seconds'].diff()
peaks = peaks.merge(y, on=peaks.index)
```

```python
peaks = peaks.drop(['key_0'], axis=1)
peaks
```

```python
plt.figure(figsize=(20, 12), dpi=100)
plt.plot(data['BVP'], label='BVP')
plt.plot(peaks['peak_points'], peaks['peak_heights'], label='peaks')
plt.plot(peaks['peak_points'], peaks['diff'], label='peak differences')
plt.legend()
sns.set(font_scale =1.5, style = 'ticks' )
plt.xlim(950,3000)
```
